```python
import numpy as np
from scipy.stats import multivariate_normal
import scipy
import matplotlib.pyplot as plt
```

```python
from GHKF import GHKF
```

Pendulum simulation

```python
n = 500
dt = 0.01
g = 9.81
R = np.array([[0.01]])
Q = np.array([[dt**3 / 3, dt**2 / 2], [dt**2 / 2, dt]]) # for q_c = 1
# Q = 0.1 * np.eye(2)
x1 = 1.5
x2 = 0
x1s = [x1]
x2s = [x2]
zs = [np.sin(x1)]
for i in range(n):
    x1_next = x1 + x2 * dt
    x2_next = x2 - g * np.sin(x1) * dt
    x1s.append(x1_next)
    x2s.append(x2_next)
    # adding process noise
    q = multivariate_normal.rvs([0,0], Q)
    x1_next += q[0]
    x2_next += q[1]
    
    z = np.sin(x1)
    # adding measurement noise
    z += multivariate_normal.rvs(0, R)
    zs.append(z)
    
    x1 = x1_next
    x2 = x2_next
```

Filter parameters:

```python
m = np.array([1.5,0]).reshape(2, 1)
#P = np.array([[1, 0], [0, 1]])
P = np.eye(2)
p = 2

def f(x, u, dt=0.1, g=9.81): # pendulum process model
    # x - np.array of shape (2, 1)
    x1_p = x[0, 0]
    x2_p = x[1, 0]
    x1 = x1_p + x2_p * dt
    x2 = x2_p - g * np.sin(x1_p) * dt
    return np.array([x1, x2]).reshape(2, 1)

def h(x): # pendulum measurement model - measuring horizontal position
    x1_p = x[0, 0]
    return np.sin(x1_p)
```

```python
ghkf = GHKF.GH_Kalman_filter(m, P, Q, R, processModel=f, measurementModel=h, degree=p)
```

```python
filter_outputs = []
filter_predictions = []
filter_covariances = []
for z in zs:
    ghkf.predict(u=0)
    filter_predictions.append(ghkf.m)
    ghkf.update(z)
    filter_outputs.append(ghkf.m)
    filter_covariances.append(ghkf.P)
```

```python
plt.figure(figsize=(10, 10), dpi=80)
t = np.linspace(0, len(zs) * dt, len(zs))
x1s_filtered = list(map(lambda x: x[0, 0], filter_outputs))
x1s_predictions = list(map(lambda x: x[0, 0], filter_predictions))
plt.plot(t, zs, 'bo')
plt.plot(t, x1s_filtered, 'r-')
plt.plot(t, x1s_predictions, 'g-')
plt.legend(['measurements', 'filtered', 'predictions'])
plt.show()
```

```python
plt.figure(figsize=(10, 10))
Ps_0_0 = list(map(lambda x: x[0, 0], filter_covariances))
plt.plot(t, Ps_0_0)
plt.show()
```

#### Pendulum 2

```python
G = np.array([0, 0.001]).reshape(2, 1)
Q = G.dot(G.T)
R = 0.02 * np.eye(2)
R_05 = scipy.linalg.sqrtm(R) 
m = np.array([0, 0]).reshape(2, 1)
P = np.eye(2)
z = np.array([0, 0]).reshape(2, 1)
nsteps = 2000;
p = 2;
dt = 0.02
sqdt = np.sqrt(dt)
def f(x, u, dt=0.02):
    x1 = x[0, 0]
    x2 = x[1, 0]
    x1_next = x1 + dt * x2
    x2_next = x2 - dt * (0.01 * x2 + 3 * np.sin(x1) + u)
    return np.array([x1_next, x2_next]).reshape(2, 1)
def h(x):
    x1 = x[0, 0]
    return np.array([np.sin(x1), -np.cos(x1)]).reshape(2, 1)
```

```python
ghkf = GHKF.GH_Kalman_filter(m, P, Q, R, processModel=f, measurementModel=h, degree=p)
```

```python
x = np.random.randn(2).reshape(2, 1)
xs = []
zs = []
ms = []
ts = []
for k in range(nsteps):
    ghkf.predict(u=0)
    # measurement
    #z = h(x) + R_05 * np.random.randn(2).reshape(2, 1)
    z = h(x) + multivariate_normal.rvs([0,0], R).reshape(2, 1)
    ghkf.update(z)
    xs.append(x)
    zs.append(z)
    ms.append(ghkf.m)
    ts.append(k * dt)
    # simulation
#     x = f(x, u=0) + G * sqdt * np.random.randn()
    x = f(x, u=0) + multivariate_normal.rvs([0,0], Q).reshape(2, 1)
```

```python
estimated_positions = list(map(lambda m: m[0, 0], ms))
true_positions = list(map(lambda x: x[0, 0], xs))
```

```python
plt.figure(figsize=(10, 10))
plt.plot(ts, estimated_positions, 'r-')
plt.plot(ts, true_positions, 'b-')
plt.legend(['estimated position', 'true position'])
plt.show()
```

```python
estimated_velocities = list(map(lambda m: m[1, 0], ms))
true_velocities = list(map(lambda x: x[1, 0], xs))
```

```python
plt.figure(figsize=(10, 10))
plt.plot(ts, estimated_velocities, 'r-')
plt.plot(ts, true_velocities, 'b-')
plt.legend(['estimated velocity', 'true velocity'])
plt.show()
```
